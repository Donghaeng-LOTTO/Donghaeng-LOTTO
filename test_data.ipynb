```python
import pandas as pd

# ⚠️ 반드시 '_by_season_' 파일인지 확인해 주세요!
df_batting_season = pd.read_csv(
    "archive\kbo_batting_stats_by_season_1982-2025.csv", encoding="utf-8-sig"
)

# 모든 컬럼 출력 설정
pd.set_option("display.max_columns", None)

# 선수 검색 및 주요 지표 확인
player_name = "손아섭"
player_stats = df_batting_season[df_batting_season["Name"] == player_name]

# 이제 'Team'과 'Age'가 정상적으로 작동합니다.
player_profile = player_stats[
    ["Year", "Team", "Age", "G", "PA", "H", "HR", "AVG", "OPS", "wRC+", "WAR"]
].sort_values(by="Year")
print(f"=== {player_name} 선수의 모든 정규시즌 기록 ===")
print(player_profile.to_string(index=False))
```
