```python
# ============================================================
# FINAL CLEAN EDA - ERROR FIXED VERSION
# 기준 파일:
# - data/processed/model_master_pa_final_label_fixed_games.csv
# - data/processed/model_dataset_ready_final_games.csv
#
# 이 코드는 라벨을 새로 만들지 않습니다.
# 이미 완성된 최종 파일을 읽어서 EDA만 수행합니다.
# ============================================================

import warnings
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

warnings.filterwarnings("ignore")

pd.set_option("display.max_columns", 160)
pd.set_option("display.max_rows", 120)
pd.set_option("display.width", 260)


# ============================================================
# 0. 경로 자동 탐색
# ============================================================

cwd = Path.cwd().resolve()

PROJECT_ROOT = None
for p in [cwd, *cwd.parents]:
    if p.name == "Donghaeng-LOTTO":
        PROJECT_ROOT = p
        break

if PROJECT_ROOT is None:
    raise FileNotFoundError("Donghaeng-LOTTO 프로젝트 루트를 찾지 못했습니다.")

PROCESSED_DIR = PROJECT_ROOT / "data" / "processed"
ANALYSIS_DIR = PROJECT_ROOT / "analysis"
OUT_DIR = ANALYSIS_DIR / "eda_outputs_final_clean"

OUT_DIR.mkdir(parents=True, exist_ok=True)

MASTER_PATH = PROCESSED_DIR / "model_master_pa_final_label_fixed_games.csv"
MODEL_PATH = PROCESSED_DIR / "model_dataset_ready_final_games.csv"

if not MASTER_PATH.exists():
    raise FileNotFoundError(f"최종 master 파일이 없습니다: {MASTER_PATH}")

if not MODEL_PATH.exists():
    raise FileNotFoundError(f"최종 model dataset 파일이 없습니다: {MODEL_PATH}")

print("PROJECT_ROOT:", PROJECT_ROOT)
print("MASTER_PATH :", MASTER_PATH)
print("MODEL_PATH  :", MODEL_PATH)
print("OUT_DIR     :", OUT_DIR)


# ============================================================
# 1. 유틸 함수
# ============================================================

TEAM_CODES = {"HH", "HT", "KT", "LG", "LT", "NC", "OB", "SK", "SS", "WO"}


def read_csv_safely(path: Path) -> pd.DataFrame:
    for enc in ["utf-8-sig", "utf-8", "cp949", "euc-kr"]:
        try:
            return pd.read_csv(path, encoding=enc, low_memory=False)
        except Exception:
            pass
    return pd.read_csv(path, low_memory=False)


def save_csv(dataframe: pd.DataFrame, name: str) -> Path:
    path = OUT_DIR / f"{name}.csv"
    dataframe.to_csv(path, index=False, encoding="utf-8-sig")
    return path


def save_fig(name: str) -> Path:
    path = OUT_DIR / name
    plt.tight_layout()
    plt.savefig(path, dpi=150, bbox_inches="tight")
    plt.show()
    plt.close()
    return path


def clean_code(series: pd.Series) -> pd.Series:
    return (
        series.astype("string")
        .str.strip()
        .str.replace(r"\.0$", "", regex=True)
        .replace(
            {
                "": pd.NA,
                "nan": pd.NA,
                "NaN": pd.NA,
                "None": pd.NA,
                "<NA>": pd.NA,
            }
        )
    )


def safe_to_numeric(dataframe: pd.DataFrame, cols: list[str]) -> pd.DataFrame:
    dataframe = dataframe.copy()
    for col in cols:
        if col in dataframe.columns:
            dataframe[col] = pd.to_numeric(dataframe[col], errors="coerce")
    return dataframe


def missing_report(dataframe: pd.DataFrame) -> pd.DataFrame:
    rows = []
    for col in dataframe.columns:
        rows.append(
            {
                "column": col,
                "dtype": str(dataframe[col].dtype),
                "missing_count": int(dataframe[col].isna().sum()),
                "missing_ratio": float(dataframe[col].isna().mean()),
                "nunique": int(dataframe[col].nunique(dropna=True)),
            }
        )

    return (
        pd.DataFrame(rows)
        .sort_values(["missing_ratio", "missing_count"], ascending=False)
        .reset_index(drop=True)
    )


def value_counts_table(dataframe: pd.DataFrame, col: str, top_n: int = 50) -> pd.DataFrame:
    vc = dataframe[col].value_counts(dropna=False).head(top_n)
    out = vc.reset_index()
    out.columns = [col, "count"]
    return out


def binary_rate_by_group(
    dataframe: pd.DataFrame,
    group_col: str,
    target_col: str,
    min_count: int = 20,
) -> pd.DataFrame:
    if group_col not in dataframe.columns or target_col not in dataframe.columns:
        return pd.DataFrame()

    temp = dataframe[[group_col, target_col]].copy()
    temp[target_col] = pd.to_numeric(temp[target_col], errors="coerce")
    temp = temp.dropna(subset=[group_col, target_col])

    if temp.empty:
        return pd.DataFrame()

    out = (
        temp.groupby(group_col, dropna=False)[target_col]
        .agg(count="size", win_rate="mean")
        .reset_index()
        .sort_values(["count", "win_rate"], ascending=[False, False])
    )

    out = out[out["count"] >= min_count].reset_index(drop=True)
    return out


def plot_bar_rate(tab: pd.DataFrame, group_col: str, title: str, filename: str, max_rows: int = 40):
    if tab.empty:
        return

    data = tab.head(max_rows).copy()

    plt.figure(figsize=(max(9, len(data) * 0.45), 5))
    plt.bar(data[group_col].astype(str), data["win_rate"])
    plt.axhline(data["win_rate"].mean(), linestyle="--", linewidth=1)
    plt.title(title)
    plt.xlabel(group_col)
    plt.ylabel("Win Rate")
    plt.ylim(0, 1)
    plt.xticks(rotation=45, ha="right")
    save_fig(filename)


def plot_line_rate(tab: pd.DataFrame, x_col: str, title: str, filename: str):
    if tab.empty:
        return

    data = tab.copy()
    data[x_col] = pd.to_numeric(data[x_col], errors="coerce")
    data = data.dropna(subset=[x_col]).sort_values(x_col)

    if data.empty:
        return

    plt.figure(figsize=(9, 5))
    plt.plot(data[x_col], data["win_rate"], marker="o")
    plt.axhline(data["win_rate"].mean(), linestyle="--", linewidth=1)
    plt.title(title)
    plt.xlabel(x_col)
    plt.ylabel("Win Rate")
    plt.ylim(0, 1)
    save_fig(filename)


# ============================================================
# 2. 데이터 불러오기
# ============================================================

master = read_csv_safely(MASTER_PATH)
model_df = read_csv_safely(MODEL_PATH)

print("\n불러오기 완료")
print("master shape:", master.shape)
print("model_df shape:", model_df.shape)

display(master.head(3))
display(model_df.head(3))


# ============================================================
# 3. 기본 타입 정리
# ============================================================

df = master.copy()

code_cols = [
    "game_id",
    "away_team_code",
    "home_team_code",
    "batting_team_code",
    "fielding_team_code",
    "winner_team_code",
    "winner_team_code_final",
    "batter_pcode_norm",
    "pitcher_pcode_norm",
]

for col in code_cols:
    if col in df.columns:
        df[col] = clean_code(df[col])

if "game_id" in df.columns:
    df["_game_date_key"] = clean_code(df["game_id"]).str.slice(0, 8)
    df["_game_date"] = pd.to_datetime(df["_game_date_key"], format="%Y%m%d", errors="coerce")
else:
    df["_game_date_key"] = pd.NA
    df["_game_date"] = pd.NaT

if "season" not in df.columns:
    df["season"] = df["_game_date"].dt.year

TARGET_COL = "target_batting_win_final"
MODEL_TARGET_COL = "target"

if TARGET_COL not in df.columns:
    if "batting_team_win_label" in df.columns:
        TARGET_COL = "batting_team_win_label"
    else:
        raise ValueError("target_batting_win_final 또는 batting_team_win_label 컬럼이 없습니다.")

if MODEL_TARGET_COL not in model_df.columns:
    raise ValueError("model_dataset_ready_final_games.csv에 target 컬럼이 없습니다.")

numeric_cols = [
    "season",
    "inning",
    "home_or_away",
    "batter_order",
    "outs_before",
    "base_state_before",
    "base1_before",
    "base2_before",
    "base3_before",
    "batting_score_before",
    "fielding_score_before",
    "batting_score_diff_before",
    "home_score_before",
    "away_score_before",
    "score_abs_diff_before",
    "score_diff_lotte_before",
    "runners_on_before",
    "is_bases_empty_before",
    "is_bases_loaded_before",
    "scoring_position_before",
    "late_clutch",
    "same_hand_matchup",
    "batter_platoon_advantage",
    "batter_pre_games_before",
    "batter_pre_avg_before",
    "batter_pre_obp_approx_before",
    "batter_pre_slg_before",
    "batter_pre_ops_before",
    "pitcher_pre_games_before",
    "pitcher_pre_era_before",
    "pitcher_pre_whip_before",
    "pitcher_pre_k9_before",
    "pitcher_pre_bb9_before",
    "batter_pre_avg_before_filled",
    "batter_pre_ops_before_filled",
    "pitcher_pre_era_before_filled",
    "pitcher_pre_whip_before_filled",
    "pitcher_pre_k9_before_filled",
    "pitcher_pre_bb9_before_filled",
    "batter_pre_avg_before_is_missing",
    "batter_pre_ops_before_is_missing",
    "pitcher_pre_era_before_is_missing",
    "pitcher_pre_whip_before_is_missing",
    "pitcher_pre_k9_before_is_missing",
    "pitcher_pre_bb9_before_is_missing",
    "home_win_label",
    "away_win_label",
    "batting_team_win_label",
    "fielding_team_win_label",
    "lotte_win_label",
    "final_win_label_lotte",
    "target_batting_win_final",
]

df = safe_to_numeric(df, numeric_cols)

# model_df는 팀 코드 두 개만 문자열로 남기고 나머지는 숫자화
model_code_cols = ["batting_team_code", "fielding_team_code"]

for col in model_code_cols:
    if col in model_df.columns:
        model_df[col] = clean_code(model_df[col])

model_numeric_cols = [c for c in model_df.columns if c not in model_code_cols]
model_df = safe_to_numeric(model_df, model_numeric_cols)


# ============================================================
# 4. 최종 품질 검증
# ============================================================

quality_rows = []

quality_rows.append({"check": "master_rows", "value": len(df)})
quality_rows.append({"check": "master_cols", "value": df.shape[1]})
quality_rows.append({"check": "model_rows", "value": len(model_df)})
quality_rows.append({"check": "model_cols", "value": model_df.shape[1]})

if "game_id" in df.columns:
    quality_rows.append({"check": "master_game_count", "value": df["game_id"].nunique()})

quality_rows.append({"check": "date_min", "value": str(df["_game_date"].min())})
quality_rows.append({"check": "date_max", "value": str(df["_game_date"].max())})

for col in ["away_team_code", "home_team_code", "batting_team_code", "fielding_team_code", "winner_team_code"]:
    if col in df.columns:
        vals = df[col].dropna().astype(str)
        bad_ratio = (~vals.isin(TEAM_CODES)).mean() if len(vals) else 0
        quality_rows.append({"check": f"{col}_bad_ratio", "value": bad_ratio})
        quality_rows.append({"check": f"{col}_unique", "value": ",".join(sorted(vals.unique()))})

for col in ["batter_name", "pitcher_name"]:
    if col in df.columns:
        quality_rows.append({"check": f"{col}_missing_count", "value": int(df[col].isna().sum())})
        quality_rows.append({"check": f"{col}_missing_ratio", "value": float(df[col].isna().mean())})

if "outs_before" in df.columns:
    bad_outs = df["outs_before"].notna() & ~df["outs_before"].isin([0, 1, 2])
    quality_rows.append({"check": "bad_outs_before_count", "value": int(bad_outs.sum())})
    quality_rows.append({"check": "bad_outs_before_ratio", "value": float(bad_outs.mean())})

quality_rows.append({"check": "duplicated_full_rows", "value": int(df.duplicated().sum())})
quality_rows.append({"check": "master_target_mean", "value": float(df[TARGET_COL].mean())})
quality_rows.append({"check": "model_target_mean", "value": float(model_df[MODEL_TARGET_COL].mean())})
quality_rows.append({"check": "model_ratio_vs_master", "value": float(len(model_df) / len(df))})

quality = pd.DataFrame(quality_rows)

print("\n[최종 품질 검증]")
display(quality)
save_csv(quality, "00_final_quality_check")


# ============================================================
# 5. 라벨 / 타깃 검증
# ============================================================

label_cols = [
    col
    for col in [
        "home_win_label",
        "away_win_label",
        "batting_team_win_label",
        "fielding_team_win_label",
        "target_batting_win_final",
        "lotte_win_label",
        "final_win_label_lotte",
    ]
    if col in df.columns
]

label_summary_rows = []

for col in label_cols:
    series = pd.to_numeric(df[col], errors="coerce")
    label_summary_rows.append(
        {
            "label_col": col,
            "non_null": int(series.notna().sum()),
            "missing": int(series.isna().sum()),
            "mean": float(series.mean()) if series.notna().any() else np.nan,
            "zero_count": int((series == 0).sum()),
            "one_count": int((series == 1).sum()),
        }
    )

label_summary = pd.DataFrame(label_summary_rows)

print("\n[라벨 요약]")
display(label_summary)
save_csv(label_summary, "01_label_summary")

for col in label_cols:
    save_csv(value_counts_table(df, col, top_n=20), f"01_value_counts_{col}")

model_target_summary = pd.DataFrame(
    [
        {
            "rows": len(model_df),
            "target_mean": float(model_df[MODEL_TARGET_COL].mean()),
            "target_0_count": int((model_df[MODEL_TARGET_COL] == 0).sum()),
            "target_1_count": int((model_df[MODEL_TARGET_COL] == 1).sum()),
            "target_0_ratio": float((model_df[MODEL_TARGET_COL] == 0).mean()),
            "target_1_ratio": float((model_df[MODEL_TARGET_COL] == 1).mean()),
        }
    ]
)

print("\n[모델 타깃 요약]")
display(model_target_summary)
save_csv(model_target_summary, "02_model_target_summary")


# ============================================================
# 6. 시즌별 요약
# ============================================================

season_summary = (
    df.groupby("season", dropna=False)
    .agg(
        rows=("game_id", "size"),
        games=("game_id", "nunique"),
        target_mean=(TARGET_COL, "mean"),
    )
    .reset_index()
    .sort_values("season")
)

if "home_win_label" in df.columns:
    home_win_season = (
        df.groupby("season", dropna=False)["home_win_label"]
        .mean()
        .reset_index(name="home_win_mean")
    )
    season_summary = season_summary.merge(home_win_season, on="season", how="left")

print("\n[시즌별 요약]")
display(season_summary)
save_csv(season_summary, "03_season_summary")

plt.figure(figsize=(9, 5))
plt.bar(season_summary["season"].astype(str), season_summary["rows"])
plt.title("Rows by Season")
plt.xlabel("Season")
plt.ylabel("Rows")
plt.xticks(rotation=45)
save_fig("03_rows_by_season.png")

plt.figure(figsize=(9, 5))
plt.plot(season_summary["season"], season_summary["target_mean"], marker="o")
plt.title("Target Mean by Season")
plt.xlabel("Season")
plt.ylabel("Target Mean")
plt.ylim(0, 1)
save_fig("04_target_mean_by_season.png")


# ============================================================
# 7. 결측치 EDA
# ============================================================

miss = missing_report(df)

print("\n[결측치 상위 50]")
display(miss.head(50))
save_csv(miss, "10_missing_report_master")

top_miss = miss.head(30)

plt.figure(figsize=(10, 9))
plt.barh(top_miss["column"][::-1], top_miss["missing_ratio"][::-1])
plt.title("Top Missing Columns - Master")
plt.xlabel("Missing Ratio")
plt.ylabel("Column")
save_fig("10_top_missing_columns_master.png")

model_miss = missing_report(model_df)

print("\n[모델 데이터 결측치]")
display(model_miss)
save_csv(model_miss, "11_missing_report_model_dataset")


# ============================================================
# 8. 모델 피처 리포트
# ============================================================

model_feature_cols = [col for col in model_df.columns if col != MODEL_TARGET_COL]

feature_report = pd.DataFrame(
    {
        "feature": model_feature_cols,
        "dtype": [str(model_df[col].dtype) for col in model_feature_cols],
        "missing_count": [int(model_df[col].isna().sum()) for col in model_feature_cols],
        "missing_ratio": [float(model_df[col].isna().mean()) for col in model_feature_cols],
        "nunique": [int(model_df[col].nunique(dropna=True)) for col in model_feature_cols],
    }
)

print("\n[모델 피처 리포트]")
display(feature_report)
save_csv(feature_report, "20_model_feature_report")


# ============================================================
# 9. 누수 위험 컬럼 후보
# ============================================================

LEAK_KEYWORDS = [
    "after",
    "result",
    "win_label",
    "winner",
    "final",
    "target",
    "score_validation",
    "score_match",
    "record_home",
    "record_away",
    "pa_result",
    "event_type",
    "result_text",
    "runner_after",
    "end_seqno",
    "result_seqno",
    "score_is_valid",
]

leak_cols = [
    col
    for col in df.columns
    if any(keyword in col.lower() for keyword in LEAK_KEYWORDS)
]

leak_report = pd.DataFrame({"leak_suspect_column": leak_cols})

print("\n[누수 위험 컬럼 후보]")
display(leak_report)
save_csv(leak_report, "30_leak_suspect_columns")


# ============================================================
# 10. 승률 패턴 EDA
# ============================================================

group_cols = [
    "season",
    "inning",
    "home_or_away",
    "is_top",
    "batting_team_code",
    "fielding_team_code",
    "batter_order",
    "outs_before",
    "base_state_before",
    "base1_before",
    "base2_before",
    "base3_before",
    "runners_on_before",
    "is_bases_empty_before",
    "is_bases_loaded_before",
    "scoring_position_before",
    "late_clutch",
    "same_hand_matchup",
    "batter_platoon_advantage",
]

for col in group_cols:
    if col not in df.columns:
        continue

    tab = binary_rate_by_group(df, col, TARGET_COL, min_count=20)

    if tab.empty:
        continue

    print(f"\n[Win Rate by {col}]")
    display(tab.head(60))
    save_csv(tab, f"40_win_rate_by_{col}")

    if tab[col].nunique() <= 60:
        plot_bar_rate(
            tab,
            col,
            title=f"{TARGET_COL} by {col}",
            filename=f"40_win_rate_by_{col}.png",
            max_rows=60,
        )


# ============================================================
# 11. 점수차 승률 EDA
# ============================================================

score_diff_cols = [
    "batting_score_diff_before",
    "score_diff_lotte_before",
    "score_abs_diff_before",
]

for col in score_diff_cols:
    if col not in df.columns:
        continue

    tab = binary_rate_by_group(df, col, TARGET_COL, min_count=20)

    if tab.empty:
        continue

    tab[col] = pd.to_numeric(tab[col], errors="coerce")
    tab = tab.dropna(subset=[col]).sort_values(col)

    print(f"\n[Win Rate by {col}]")
    display(tab.head(100))
    save_csv(tab, f"50_win_rate_by_{col}")

    plot_line_rate(
        tab,
        col,
        title=f"{TARGET_COL} by {col}",
        filename=f"50_win_rate_by_{col}.png",
    )


# ============================================================
# 12. 수치형 상관 분석
# ============================================================

numeric_feature_cols = [
    col
    for col in model_feature_cols
    if col not in ["batting_team_code", "fielding_team_code"]
]

corr_df = model_df[numeric_feature_cols + [MODEL_TARGET_COL]].copy()
corr_df = safe_to_numeric(corr_df, corr_df.columns.tolist())

corr = corr_df.corr(numeric_only=True)[MODEL_TARGET_COL].drop(MODEL_TARGET_COL, errors="ignore")

corr_table = (
    corr.rename("corr_with_target")
    .reset_index()
    .rename(columns={"index": "feature"})
)

corr_table["abs_corr"] = corr_table["corr_with_target"].abs()
corr_table["non_null_count"] = [
    int(corr_df[col].notna().sum()) for col in corr_table["feature"]
]
corr_table = corr_table.sort_values("abs_corr", ascending=False).reset_index(drop=True)

print("\n[모델 피처 상관 분석]")
display(corr_table)
save_csv(corr_table, "60_model_numeric_feature_corr")

top_corr = corr_table.head(25)

plt.figure(figsize=(10, 8))
plt.barh(top_corr["feature"][::-1], top_corr["corr_with_target"][::-1])
plt.title("Top Correlations with Target")
plt.xlabel("Correlation")
plt.ylabel("Feature")
save_fig("60_top_corr_with_target.png")


# ============================================================
# 13. 주요 피처 describe
# ============================================================

desc_cols = [
    col for col in numeric_feature_cols if col in model_df.columns
]

desc = (
    model_df[desc_cols]
    .describe()
    .T
    .reset_index()
    .rename(columns={"index": "feature"})
)

print("\n[모델 수치형 피처 describe]")
display(desc)
save_csv(desc, "70_model_numeric_feature_describe")


# ============================================================
# 14. 롯데 전용 EDA
# ============================================================

lotte_mask = pd.Series(False, index=df.index)

if "batting_team_code" in df.columns:
    lotte_mask = lotte_mask | df["batting_team_code"].eq("LT")

if "fielding_team_code" in df.columns:
    lotte_mask = lotte_mask | df["fielding_team_code"].eq("LT")

lotte = df[lotte_mask].copy()

LOTTE_TARGET = None

for col in ["final_win_label_lotte", "lotte_win_label"]:
    if col in lotte.columns and lotte[col].notna().sum() > 0:
        LOTTE_TARGET = col
        break

lotte_overview = pd.DataFrame(
    [
        {
            "lotte_rows": len(lotte),
            "lotte_games": lotte["game_id"].nunique() if "game_id" in lotte.columns else np.nan,
            "lotte_batting_rows": int(df["batting_team_code"].eq("LT").sum()) if "batting_team_code" in df.columns else np.nan,
            "lotte_fielding_rows": int(df["fielding_team_code"].eq("LT").sum()) if "fielding_team_code" in df.columns else np.nan,
            "lotte_target": LOTTE_TARGET,
            "lotte_target_non_null": int(lotte[LOTTE_TARGET].notna().sum()) if LOTTE_TARGET else 0,
            "lotte_target_mean": float(lotte[LOTTE_TARGET].mean()) if LOTTE_TARGET else np.nan,
        }
    ]
)

print("\n[롯데 데이터 요약]")
display(lotte_overview)
save_csv(lotte_overview, "80_lotte_overview")

if LOTTE_TARGET is not None:
    lotte_group_cols = [
        "season",
        "inning",
        "home_or_away",
        "batting_team_code",
        "fielding_team_code",
        "outs_before",
        "base_state_before",
        "batter_order",
        "batting_score_diff_before",
        "score_diff_lotte_before",
    ]

    for col in lotte_group_cols:
        if col not in lotte.columns:
            continue

        tab = binary_rate_by_group(lotte, col, LOTTE_TARGET, min_count=10)

        if tab.empty:
            continue

        print(f"\n[Lotte Win Rate by {col}]")
        display(tab.head(60))
        save_csv(tab, f"81_lotte_win_rate_by_{col}")

        if col in ["batting_score_diff_before", "score_diff_lotte_before"]:
            plot_line_rate(
                tab,
                col,
                title=f"{LOTTE_TARGET} by {col}",
                filename=f"81_lotte_win_rate_by_{col}.png",
            )
        elif tab[col].nunique() <= 60:
            plot_bar_rate(
                tab,
                col,
                title=f"{LOTTE_TARGET} by {col}",
                filename=f"81_lotte_win_rate_by_{col}.png",
                max_rows=60,
            )


# ============================================================
# 15. 베이스라인 통계
# ============================================================

baseline_rows = []

baseline_rows.append(
    {
        "baseline": "overall_target_mean",
        "description": "전체 평균 승률",
        "value": float(model_df[MODEL_TARGET_COL].mean()),
    }
)

if "batting_score_diff_before" in df.columns:
    temp = df[[TARGET_COL, "batting_score_diff_before"]].dropna().copy()

    temp["lead_state"] = np.where(
        temp["batting_score_diff_before"] > 0,
        "leading",
        np.where(temp["batting_score_diff_before"] < 0, "trailing", "tied"),
    )

    lead_tab = (
        temp.groupby("lead_state")[TARGET_COL]
        .agg(count="size", win_rate="mean")
        .reset_index()
    )

    save_csv(lead_tab, "90_baseline_win_rate_by_lead_state")

    for _, row in lead_tab.iterrows():
        baseline_rows.append(
            {
                "baseline": f"lead_state_{row['lead_state']}",
                "description": "현재 공격팀 점수차 상태별 평균 승률",
                "value": float(row["win_rate"]),
            }
        )

baseline_report = pd.DataFrame(baseline_rows)

print("\n[베이스라인 통계]")
display(baseline_report)
save_csv(baseline_report, "90_baseline_report")


# ============================================================
# 16. 최종 텍스트 리포트 저장
# ============================================================

report_lines = []

report_lines.append("=" * 80)
report_lines.append("FINAL CLEAN EDA REPORT")
report_lines.append("=" * 80)
report_lines.append("")
report_lines.append(f"MASTER_PATH: {MASTER_PATH}")
report_lines.append(f"MODEL_PATH: {MODEL_PATH}")
report_lines.append(f"OUT_DIR: {OUT_DIR}")
report_lines.append("")
report_lines.append("[DATA]")
report_lines.append(f"master rows: {len(df):,}")
report_lines.append(f"master cols: {df.shape[1]:,}")
report_lines.append(f"model rows: {len(model_df):,}")
report_lines.append(f"model cols: {model_df.shape[1]:,}")

if "game_id" in df.columns:
    report_lines.append(f"games: {df['game_id'].nunique():,}")

report_lines.append(f"date_min: {df['_game_date'].min()}")
report_lines.append(f"date_max: {df['_game_date'].max()}")
report_lines.append("")
report_lines.append("[TARGET]")
report_lines.append(f"master target col: {TARGET_COL}")
report_lines.append(f"master target mean: {df[TARGET_COL].mean():.6f}")
report_lines.append(f"model target mean: {model_df[MODEL_TARGET_COL].mean():.6f}")
report_lines.append(f"model target 1 count: {int((model_df[MODEL_TARGET_COL] == 1).sum()):,}")
report_lines.append(f"model target 0 count: {int((model_df[MODEL_TARGET_COL] == 0).sum()):,}")
report_lines.append("")
report_lines.append("[SEASON SUMMARY]")

for _, row in season_summary.iterrows():
    report_lines.append(
        f"season={row['season']}, rows={int(row['rows'])}, games={int(row['games'])}, target_mean={row['target_mean']:.6f}"
    )

report_lines.append("")
report_lines.append("[TOP CORRELATIONS]")

for _, row in corr_table.head(15).iterrows():
    report_lines.append(f"{row['feature']}: corr={row['corr_with_target']:.6f}")

report_lines.append("")
report_lines.append("[LOTTE]")

for _, row in lotte_overview.iterrows():
    for col in lotte_overview.columns:
        report_lines.append(f"{col}: {row[col]}")

report_lines.append("")
report_lines.append("[OUTPUT FILES]")
report_lines.append("00_final_quality_check.csv")
report_lines.append("01_label_summary.csv")
report_lines.append("02_model_target_summary.csv")
report_lines.append("03_season_summary.csv")
report_lines.append("10_missing_report_master.csv")
report_lines.append("20_model_feature_report.csv")
report_lines.append("30_leak_suspect_columns.csv")
report_lines.append("40_win_rate_by_*.csv/png")
report_lines.append("50_win_rate_by_*.csv/png")
report_lines.append("60_model_numeric_feature_corr.csv/png")
report_lines.append("80_lotte_overview.csv")
report_lines.append("90_baseline_report.csv")
report_lines.append("")
report_lines.append("=" * 80)
report_lines.append("END")
report_lines.append("=" * 80)

REPORT_PATH = OUT_DIR / "00_final_eda_report_for_chatgpt.txt"
REPORT_PATH.write_text("\n".join(report_lines), encoding="utf-8-sig")

print("\n텍스트 리포트 저장:", REPORT_PATH)


# ============================================================
# 17. 최종 완료 요약
# ============================================================

final_summary = pd.DataFrame(
    [
        {
            "master_rows": len(df),
            "master_cols": df.shape[1],
            "model_rows": len(model_df),
            "model_cols": model_df.shape[1],
            "model_ratio": len(model_df) / len(df),
            "master_target_mean": df[TARGET_COL].mean(),
            "model_target_mean": model_df[MODEL_TARGET_COL].mean(),
            "output_dir": str(OUT_DIR),
            "report_path": str(REPORT_PATH),
        }
    ]
)

print("\n" + "=" * 80)
print("FINAL CLEAN EDA COMPLETE")
print("=" * 80)

display(final_summary)
save_csv(final_summary, "99_final_summary")
```
