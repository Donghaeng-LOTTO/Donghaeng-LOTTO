```python
import pandas as pd

pa = pd.read_csv("data/processed/plate_appearances.csv")

target_game = "20080601HTOB0"
g = pa[pa["game_id"] == target_game].copy()

score_cols = [
    "game_id",
    "inning",
    "home_or_away",
    "result_text",
    "runner_before_result_text",
    "runner_after_result_text",
    "away_score_before",
    "home_score_before",
    "parse_warning",
]

existing = [c for c in score_cols if c in g.columns]

# 홈인 여부를 검사할 텍스트 컬럼만 선택
text_cols = [
    "result_text",
    "runner_before_result_text",
    "runner_after_result_text",
    "change_text",
]

text_cols = [c for c in text_cols if c in g.columns]

# NaN 제거 + 문자열 변환 후 합치기
homein_text = (
    g[text_cols]
    .fillna("")
    .astype(str)
    .agg(" ".join, axis=1)
)

scoring_rows = g[homein_text.str.contains("홈인", na=False)]

print(scoring_rows[existing].head(50).to_string(index=False))
print(g["home_or_away"].value_counts(dropna=False))
print(g["home_or_away"].map(lambda x: (x, type(x))).head(10))
```

```python
import pandas as pd

pa = pd.read_csv("data/processed/plate_appearances.csv")

target_game = "20080601HTOB0"
g = pa[pa["game_id"] == target_game].copy()

text_cols = [
    "result_text",
    "runner_before_result_text",
    "runner_after_result_text",
]

for c in text_cols:
    if c not in g.columns:
        g[c] = ""

g["all_text"] = (
    g[text_cols]
    .fillna("")
    .astype(str)
    .agg(" | ".join, axis=1)
)

away_scoring = g[
    (g["home_or_away"].astype(str).isin(["0", "0.0"])) &
    (g["all_text"].str.contains("홈인", na=False))
]

cols = [
    "inning",
    "home_or_away",
    "away_score_before",
    "home_score_before",
    "away_score_after",
    "home_score_after",
    "result_text",
    "runner_before_result_text",
    "runner_after_result_text",
    "parse_warning",
]

print(away_scoring[cols].to_string(index=False))
```
